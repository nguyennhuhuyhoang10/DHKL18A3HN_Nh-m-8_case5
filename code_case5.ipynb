```python
import numpy as np
#PHẦN 7 Boolean mask – so sánh hai máy chủ 
# Load dữ liệu từ file CSV
s1 = np.genfromtxt("server1_log.csv", delimiter=",", dtype=str, skip_header=1)
s2 = np.genfromtxt("server2_log.csv", delimiter=",", dtype=str, skip_header=1)
# tách cột từ mảng s1 thành mảng riêng và chuyển dữ liệu sang kiểu float,int,...
time1 = s1[:,0].astype("datetime64[m]")
cpu1  = s1[:,1].astype(float)
ram1  = s1[:,2].astype(float)
err1  = s1[:,3].astype(int)
# tách cột từ mảng s2 thành mảng riêng và chuyển dữ liệu sang kiểu float,int,...
time2 = s2[:,0].astype("datetime64[m]")
cpu2  = s2[:,1].astype(float)
ram2  = s2[:,2].astype(float)
err2  = s2[:,3].astype(int)

# Tạo mask: CPU_s1 >= CPU_s2 + 10
mask_cpu_diff = cpu1 >= (cpu2 + 10)

# Tạo masked array
masked_time = np.ma.masked_array(time1, mask=~mask_cpu_diff)
masked_cpu1 = np.ma.masked_array(cpu1, mask=~mask_cpu_diff)
masked_cpu2 = np.ma.masked_array(cpu2, mask=~mask_cpu_diff)

# Duyệt bằng index và trả về danh sách khung giờ
for i in range(len(masked_time)):
    if not masked_time.mask[i]:  # nếu phần tử không bị mask
        print(f"{masked_time[i]} | CPU Server 1: {masked_cpu1[i]}% | CPU Server 2: {masked_cpu2[i]}%")
```

```python
# 8. FANCY INDEXING – TOP-5 GIỜ CPU CAO NHẤT

top5_s1 = np.argsort(cpu1)[-5:]
top5_s2 = np.argsort(cpu2)[-5:]

print("\nTop 5 CPU cao nhất S1:")
for i in top5_s1:
    print(time1[i], cpu1[i], ram1[i], err1[i])

print("\nTop 5 CPU cao nhất S2:")
for i in top5_s2:
    print(time2[i], cpu2[i], ram2[i], err2[i])
```

```python
# 9.Broadcasting – so sánh hiệu năng giữa hai máy chủ 

diff_cpu = cpu1 - cpu2
diff_ram = ram1 - ram2

idx_max_cpu1_vs2 = np.argmax(diff_cpu)
idx_max_ram2_vs1 = np.argmax(-diff_ram)

print("\nGiờ CPU S1 > S2 nhiều nhất:", time1[idx_max_cpu1_vs2])
print("Chênh lệch:", diff_cpu[idx_max_cpu1_vs2])

print("\nGiờ RAM S2 > S1 nhiều nhất:", time1[idx_max_ram2_vs1])
print("Chênh lệch:", -diff_ram[idx_max_ram2_vs1])
```

```python
#PHẦN 10 np.diff – phân tích biến động theo giờ 

# 1. Tách thời gian và CPU đúng cấu trúc structured-array
time1 = dataA['time']
time2 = dataB['time']

cpu1 = dataA['cpu_percent'].astype(float)
cpu2 = dataB['cpu_percent'].astype(float)

# 2. Tính biến động CPU mỗi giờ
diff_cpu1 = np.diff(cpu1)
diff_cpu2 = np.diff(cpu2)

# 3. In kết quả
print("=== Biến động CPU Server 1 ===")
for i, delta in enumerate(diff_cpu1):
    print(f"{time1[i]} → {time1[i+1]} : ΔCPU = {delta:.2f}")

print("\n=== Biến động CPU Server 2 ===")
for i, delta in enumerate(diff_cpu2):
    print(f"{time2[i]} → {time2[i+1]} : ΔCPU = {delta:.2f}")
```

```python
# Đặt ngưỡng biến động 
threshold = 15

spike1 = np.where(np.abs(diff_cpu1) >= threshold)[0]
spike2 = np.where(np.abs(diff_cpu2) >= threshold)[0]

print("=== Server 1 – Biến động CPU mạnh (|ΔCPU| ≥ 15%) ===")
for idx in spike1:
    print(f"{time1[idx]} → {time1[idx+1]} : ΔCPU = {diff_cpu1[idx]:.2f}")

print("\n=== Server 2 – Biến động CPU mạnh (|ΔCPU| ≥ 15%) ===")
for idx in spike2:
    print(f"{time2[idx]} → {time2[idx+1]} : ΔCPU = {diff_cpu2[idx]:.2f}")
```

```python
#tính biến động error_count giữ các giờ liên tiếp
err1 = dataA['error_count'].astype(int)
err2 = dataB['error_count'].astype(int)
diff_err1 = np.diff(err1)
diff_err2 = np.diff(err2)
# đặt ngưỡng đột biến
threshold_err = 5

spike_err1 = np.where(np.abs(diff_err1) >= threshold_err)[0]
spike_err2 = np.where(np.abs(diff_err2) >= threshold_err)[0]

print("=== dataA – Error count tăng/giảm đột biến (|Δerror| ≥ 5) ===")
for idx in spike_err1:
    print(f"{time1[idx]} → {time1[idx+1]} : Δerror = {diff_err1[idx]}")

print("\n=== dataB – Error count tăng/giảm đột biến (|Δerror| ≥ 5) ===")
for idx in spike_err2:
    print(f"{time2[idx]} → {time2[idx+1]} : Δerror = {diff_err2[idx]}")
```

```python
#PHẦN 11 Thống kê tổng hợp 
#chuyển sang dang float/int
cpu1 = dataA['cpu_percent'].astype(float)
ram1 = dataA['ram_percent'].astype(float)
err1 = dataA['error_count'].astype(int)

cpu2 = dataB['cpu_percent'].astype(float)
ram2 = dataB['ram_percent'].astype(float)
err2 = dataB['error_count'].astype(int)
#tính các thống kê cơ bản
def print_stats(name, arr):
    print(f"\n=== {name} ===")
    print("Mean :", arr.mean())
    print("Std  :", arr.std())
    print("Max  :", arr.max())
    print("Min  :", arr.min())

print("### dataA ###")
print_stats("CPU", cpu1)
print_stats("RAM", ram1)
print_stats("Error_count", err1)

print("\n### dataB ###")
print_stats("CPU", cpu2)
print_stats("RAM", ram2)
print_stats("Error_count", err2)
```

```python

#hàm in index và thời gian max/min
def print_extreme(name, time_arr, arr):
    idx_max = np.argmax(arr)
    idx_min = np.argmin(arr)
    print(f"\n{name}:")
    print(f"Max = {arr[idx_max]} tại giờ {time_arr[idx_max]} (index {idx_max})")
    print(f"Min = {arr[idx_min]} tại giờ {time_arr[idx_min]} (index {idx_min})")

print("### SERVER 1 ###")
print_extreme("CPU", time1, cpu1)
print_extreme("RAM", time1, ram1)
print_extreme("Error_count", time1, err1)

print("\n### SERVER 2 ###")
print_extreme("CPU", time2, cpu2)
print_extreme("RAM", time2, ram2)
print_extreme("Error_count", time2, err2)
```

```python
#Phần 12
# 1) TẠO NHÃN LABEL
# tạo nhãn theo đúng kích thước cpu1 (24 dòng)
label = np.zeros_like(cpu1, dtype=int)   # 0 = bình thường

# ---- LEVEL 1: cảnh báo (CPU hoặc RAM > 80) ----
mask_warn = (
    (cpu1 > 80) | (ram1 > 80) |
    (cpu2 > 80) | (ram2 > 80)
)
label[mask_warn] = 1
# ---- LEVEL 2: nguy cơ cao (CPU/RAM > 90 hoặc lỗi tăng đột biến) ----
mask_high_load = (
    (cpu1 > 90) | (ram1 > 90) |
    (cpu2 > 90) | (ram2 > 90)
)

# diff error để phát hiện lỗi tăng mạnh
diff_err1 = np.diff(err1, prepend=err1[0])
diff_err2 = np.diff(err2, prepend=err2[0])

# ngưỡng lỗi tăng đột biến
thr1 = diff_err1.mean() + 2 * diff_err1.std()
thr2 = diff_err2.mean() + 2 * diff_err2.std()

mask_err_spike = (diff_err1 > thr1) | (diff_err2 > thr2)

# gộp tất cả điều kiện level 2
mask_level2 = mask_high_load | mask_err_spike
label[mask_level2] = 2

# 2) TẠO BẢNG TỔNG HỢP 2D
# chuyển thời gian sang chuỗi để ghép cột dễ hơn
time_str = time1.astype(str)

output = np.column_stack([
    time_str,        # 1. thời gian
    cpu1,            # 2. cpu_s1
    ram1,            # 3. ram_s1
    err1,            # 4. err_s1
    cpu2,            # 5. cpu_s2
    ram2,            # 6. ram_s2
    err2,            # 7. err_s2
    cpu1 - cpu2,     # 8. diff_cpu
    ram1 - ram2,     # 9. diff_ram
    label            # 10. nhãn
])
print(cpu1.shape)
print(cpu2.shape)
print(label.shape)

# 3) LƯU FILE CSV
np.savetxt(
    "ket_qua_case_5.csv",
    output,
    delimiter=",",
    fmt="%s",
    header="time,cpu_s1,ram_s1,err_s1,cpu_s2,ram_s2,err_s2,diff_cpu,diff_ram,label",
    comments=""
)

print(">>> Đã lưu file thành công: ket_qua_case_5.csv")

```
